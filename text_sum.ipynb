Prompt Template Text Summarization

```python
from langchain.chains import LLMChain
from langchain import PromptTemplate

generictemplate=""" 
Write a summary of the following speech
speech:{speech}
Translate the precise summary to {language}
"""
prompt=PromptTemplate(
    input_variable=['speech','language'],
    template=generictemplate
)
print(prompt)
```

```python
complete_prompt=prompt.format(speech=speech,language="French")
print(complete_prompt)
#the token will be increacsed 
```

```python
llm_chain=LLMChain(llm=model,prompt=prompt)
summary=llm_chain.run({"speech":speech,"language":"Urdu"})
print(summary)
```

```python
#there is the better way it as there are lot of big data and we have a token 9000 so for that we will be using the 
#Stuff Documnet Chain
#One of the basic techinque 
#pdf ->prompt template -> LLM -> O/P
#when you have 10 pdf it combines and then it is combined and send to the prompt template ,by combine mean we get all the text.
#now here the problem arise which is the documnet is larger and the token size of model is confined .
#notice that if it not worked in fits in LLM context window 
```

```python
#Map Reduce Summarization 
#documnets divided in to the chunks and then send to the prompt and then send to the llm model and thus we get sum1,sum2,sum3 and then we combine it and then it again prompt template and 
#then we will get the final summary 
```

```python
from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader("apjspeech.pdf")
page=loader.load_and_split()
print(len(page)) #documnet divided we can clearly see that 
```

```python
template="""
Write concise and short summary of the following speech,
Speech:{text}

"""
prompt=PromptTemplate(input_variables=['text'],template=template)
```

```python
from langchain.chains.summarize import load_summarize_chain
chain=load_summarize_chain(model,chain_type='stuff',prompt=prompt,verbose=True)
chain.run(page)
```

```python
#now we will do summarization using map
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_spiltter=RecursiveCharacterTextSplitter(chunk_size=2000,chunk_overlap=100).split_documents(page)
print(len(text_spiltter))
```

```python
#now we will take one by one documnet and then forward it to the prompt and get a summary 
chunk_prompt="""
Please summarize the below speech :
Speech:{text}
summary:
"""
map_prompt_template=PromptTemplate(input_variables=['text'],template=chunk_prompt)
```

```python
#the final template prompt 
final_prompt='''
Provide the final Summary of the entire speech with these important Points:
Add a Motivation Title ,start the precise summary with an introducation and provide the summary in numbers and points for the speech
Speech:{text}
'''
final_prompt_template=PromptTemplate(input_variables=['text'],template=final_prompt)
```

```python
#now apply it bu this is for the chunks of the documnets 
summary_chain=load_summarize_chain(
    llm=model,
    chain_type="map_reduce",
    map_prompt=map_prompt_template,
    combine_prompt=final_prompt_template,
    verbose=True
)
```

```python
output=summary_chain.run(text_spiltter)
```

```python
#Refine chain summarization ,it is same as stuff and map chain just an slight modification  we will be taking the referance of the previous summary
summary_chain=load_summarize_chain(
    llm=model,
    chain_type="refine",
    verbose=True
)
output=summary_chain.run(text_spiltter)
```
